### **`Smartphone Sales`**

Business problem untuk kasus ini adalah : 

- Brand mana yang memiliki penjualan paling banyak ?
- Warna apa yang paling banyak diminati oleh Customer ? 
- Berapa rata-rata harga smartphone yang ada di dalam data penjualan tersebut ?
- Smartphone mana yang paling laku (top-1 product)?
- Tahun rilis mana yang memiliki jumlah produk terbanyak?

Selain 3 business problems di atas, buat minimal 2 buah business problem yang harus kamu cari jawabannya.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("./file/SmartphonesDataset.csv")

print("Preview dataset:")
display(df.head())

# =========================
# 1. Brand dengan jumlah produk terbanyak
# =========================
brand_counts = df["Brand"].value_counts()
top_brand = brand_counts.head(1)

print("\nBrand dengan jumlah produk terbanyak:")
print(top_brand)

# Visualisasi
plt.figure(figsize=(8,5))
brand_counts.plot(kind="bar", color="skyblue")
plt.title("Jumlah Produk per Brand")
plt.ylabel("Jumlah Produk")
plt.xticks(rotation=45, ha="right")
plt.show()

# =========================
# 2. Warna paling diminati
# =========================
if "Color" in df.columns:
    top_color = df["Color"].value_counts().head(1)
    print("\nWarna paling diminati customer:")
    print(top_color)

    # Visualisasi
    plt.figure(figsize=(6,6))
    df["Color"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=140)
    plt.title("Distribusi Warna Smartphone")
    plt.ylabel("")
    plt.show()

# =========================
# 3. Rata-rata harga smartphone
# =========================
if "Price" in df.columns:
    # pastikan Price berupa angka
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
    avg_price = df["Price"].mean()
    print("\nRata-rata harga smartphone:", round(avg_price,2))

# =========================
# 4. Smartphone terlaris (pakai jumlah kemunculan Model, karena kolom 'Sales' tidak ada)
# =========================
if "Model" in df.columns:
    top_product = df["Model"].value_counts().head(1)
    print("\nSmartphone terlaris (paling banyak muncul di dataset):")
    print(top_product)

# =========================
# 5. Tahun rilis dengan jumlah produk terbanyak
# =========================
if "Release_Year" in df.columns:
    top_year = df["Release_Year"].value_counts().head(1)
    print("\nTahun dengan jumlah produk terbanyak:")
    print(top_year)

    # Visualisasi
    plt.figure(figsize=(8,5))
    df["Release_Year"].value_counts().sort_index().plot(kind="bar", color="orange")
    plt.title("Jumlah Produk per Tahun")
    plt.ylabel("Jumlah Produk")
    plt.xlabel("Tahun Rilis")
    plt.show()
else:
    print("\nKolom Release_Year tidak ada di dataset.")
```

### **`Amazon Top Selling Books`**

Business problem untuk kasus ini adalah : 

- Penulis mana yang memiliki review paling banyak ?
- Penulis mana yang memiliki score rating tertinggi ? 
- Genre mana yang paling banyak muncul di dalam dataset tersebut ?
- Buku mana yang memiliki rating tertinggi dengan jumlah review terbanyak?
- Tahun berapa yang memiliki jumlah buku terbit paling banyak?

Selain 3 business problems di atas, buat minimal 2 buah business problem yang harus kamu cari jawabannya.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("./file/AmazonTopSellingBooks.csv")

# Melihat struktur data
print("Preview Data:")
display(df.head())

print("\nInformasi dataset:")
print(df.info())

# 1. Penulis dengan review terbanyak
author_most_reviews = df.groupby("Author")["Reviews"].sum().sort_values(ascending=False).head(1)
print("\nPenulis dengan review terbanyak:")
display(author_most_reviews)

# 2. Penulis dengan rating tertinggi
author_highest_rating = df.groupby("Author")["User Rating"].mean().sort_values(ascending=False).head(1)
print("\nPenulis dengan rating tertinggi:")
display(author_highest_rating)

# 3. Genre yang paling banyak muncul
most_common_genre = df["Genre"].value_counts().head(1)
print("\nGenre paling banyak muncul:")
display(most_common_genre)

# 4. Buku dengan rating tertinggi & jumlah review terbanyak
top_book = df.sort_values(by=["User Rating", "Reviews"], ascending=[False, False]).head(1)
print("\nBuku dengan rating tertinggi & review terbanyak:")
display(top_book[["Name", "Author", "User Rating", "Reviews"]])

# 5. Tahun dengan jumlah buku terbit terbanyak
most_books_year = df["Year"].value_counts().head(1)
print("\nTahun dengan jumlah buku terbit terbanyak:")
display(most_books_year)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("./file/AmazonTopSellingBooks.csv")

# =========================
# 1. Penulis dengan review terbanyak
# =========================
author_reviews = df.groupby("Author")["Reviews"].sum().sort_values(ascending=False)
top_author_reviews = author_reviews.head(1)

print("Penulis dengan review terbanyak:")
print(top_author_reviews)

# Visualisasi
plt.figure(figsize=(10,5))
author_reviews.head(10).plot(kind="bar", color="skyblue")
plt.title("Top 10 Penulis dengan Review Terbanyak")
plt.ylabel("Jumlah Review")
plt.xticks(rotation=45, ha="right")
plt.show()

# =========================
# 2. Penulis dengan rating tertinggi
# =========================
author_ratings = df.groupby("Author")["User Rating"].mean().sort_values(ascending=False)
top_author_rating = author_ratings.head(1)

print("\nPenulis dengan rating tertinggi:")
print(top_author_rating)

# Visualisasi
plt.figure(figsize=(10,5))
author_ratings.head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Penulis dengan Rating Tertinggi (Rata-rata)")
plt.ylabel("Rata-rata Rating")
plt.xticks(rotation=45, ha="right")
plt.show()

# =========================
# 3. Genre paling banyak muncul
# =========================
genre_counts = df["Genre"].value_counts()
top_genre = genre_counts.head(1)

print("\nGenre paling banyak muncul:")
print(top_genre)

# Visualisasi (pie chart)
plt.figure(figsize=(6,6))
genre_counts.plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=["#66b3ff","#99ff99"])
plt.title("Distribusi Genre")
plt.ylabel("")
plt.show()

# =========================
# 4. Buku dengan rating tertinggi + review terbanyak
# =========================
top_book = df.sort_values(by=["User Rating", "Reviews"], ascending=[False, False]).head(1)

print("\nBuku dengan rating tertinggi & review terbanyak:")
print(top_book[["Name", "Author", "User Rating", "Reviews"]])

# =========================
# 5. Tahun dengan jumlah buku terbit terbanyak
# =========================
year_counts = df["Year"].value_counts()
top_year = year_counts.head(1)

print("\nTahun dengan jumlah buku terbit terbanyak:")
print(top_year)

# Visualisasi
plt.figure(figsize=(10,5))
year_counts.plot(kind="bar", color="green")
plt.title("Jumlah Buku Terbit per Tahun")
plt.ylabel("Jumlah Buku")
plt.xlabel("Tahun")
plt.show()
```
